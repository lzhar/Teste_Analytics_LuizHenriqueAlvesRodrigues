### üßÆ C√°lculo de Produto Mais Vendido por M√™s  

Essa √© a c√©lula respons√°vel por calcular **qual produto foi mais vendido em cada m√™s** üìä.  

Ela retorna **uma lista**, que √© composta de outras listas üóÇÔ∏è, e cada uma guarda um **dicion√°rio (`dict`)** com os seguintes valores:  

- üÜî **id**  
- üè∑Ô∏è **nome_produto**  
- üóÇÔ∏è **produto_categoria**  
- üì¶ **qtd**  
- üí∞ **preco**  
- üìÖ **mes**

```python

class Calculos:

    def __init__(self, classe_base):
        self.classe_base = classe_base

    def calcular_total_produtos(self):
        produtos = []
        lista_de_dados = self.classe_base.retornar_produtos_sem_duplicata()

        for mes in lista_de_dados:
            for registro in mes:
                if registro["id"] in produtos:
                    continue
                else:
                    produtos.append(registro["id"])


        return produtos


    def calcular_produto_mais_vendido_de_cada_mes(self):
        produtos = sorted(self.calcular_total_produtos())

        produto_mais_ven_jan = []
        produto_mais_ven_fev = []
        produto_mais_ven_marco = []
        produto_mais_ven_abril = []
        produto_mais_ven_maio = []
        produto_mais_ven_junho = []
        produto_mais_ven_julho = []
        produto_mais_ven_agosto = []
        produto_mais_ven_setembro = []
        produto_mais_ven_out = []
        produto_mais_ven_nov = []
        produto_mais_ven_dezem = []



        produtos_sem_duplicata = self.classe_base.retornar_produtos_sem_duplicata()

        for item in produtos_sem_duplicata:
            for registro in item:
                mes = registro["mes"]
                
                for produto in item:
                    if mes == "janeiro":
                        if len(produto_mais_ven_jan) == 0: 
                            produto_mais_ven_jan.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_jan[0]["qtd"]:
                            del(produto_mais_ven_jan[0])
                            produto_mais_ven_jan.append(produto)
                    
                    if mes == "fevereiro":
                        if len(produto_mais_ven_fev) == 0: 
                            produto_mais_ven_fev.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_fev[0]["qtd"]:
                            del(produto_mais_ven_fev[0])
                            produto_mais_ven_fev.append(produto)
                    
                    if mes == "marco":
                        if len(produto_mais_ven_marco) == 0: 
                            produto_mais_ven_marco.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_marco[0]["qtd"]:
                            del(produto_mais_ven_marco[0])
                            produto_mais_ven_marco.append(produto)
                    
                    if mes == "abril":
                        if len(produto_mais_ven_abril) == 0: 
                            produto_mais_ven_abril.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_abril[0]["qtd"]:
                            del(produto_mais_ven_abril[0])
                            produto_mais_ven_abril.append(produto)
                    
                    if mes == "maio":
                        if len(produto_mais_ven_maio) == 0: 
                            produto_mais_ven_maio.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_maio[0]["qtd"]:
                            del(produto_mais_ven_maio[0])
                            produto_mais_ven_maio.append(produto)

                    if mes == "junho":
                        if len(produto_mais_ven_junho) == 0: 
                            produto_mais_ven_junho.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_junho[0]["qtd"]:
                            del(produto_mais_ven_junho[0])
                            produto_mais_ven_junho.append(produto)

                    if mes == "julho":
                        if len(produto_mais_ven_julho) == 0: 
                            produto_mais_ven_julho.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_julho[0]["qtd"]:
                            del(produto_mais_ven_julho[0])
                            produto_mais_ven_julho.append(produto)
                    
                    if mes == "agosto":
                        if len(produto_mais_ven_agosto) == 0: 
                            produto_mais_ven_agosto.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_agosto[0]["qtd"]:
                            del(produto_mais_ven_agosto[0])
                            produto_mais_ven_agosto.append(produto)

                    if mes == "setembro":
                        if len(produto_mais_ven_setembro) == 0: 
                            produto_mais_ven_setembro.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_setembro[0]["qtd"]:
                            del(produto_mais_ven_setembro[0])
                            produto_mais_ven_setembro.append(produto)

                    if mes == "outubro":
                        if len(produto_mais_ven_out) == 0: 
                            produto_mais_ven_out.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_out[0]["qtd"]:
                            del(produto_mais_ven_out[0])
                            produto_mais_ven_out.append(produto)

                    if mes == "novembro":
                        if len(produto_mais_ven_nov) == 0: 
                            produto_mais_ven_nov.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_nov[0]["qtd"]:
                            del(produto_mais_ven_nov[0])
                            produto_mais_ven_nov.append(produto)

                    if mes == "dezembro":
                        if len(produto_mais_ven_dezem) == 0: 
                            produto_mais_ven_dezem.append(produto)
                            break
                        if produto["qtd"] > produto_mais_ven_dezem[0]["qtd"]:
                            del(produto_mais_ven_dezem[0])
                            produto_mais_ven_dezem.append(produto)

        lista_de_todos_mais_vendidos = [produto_mais_ven_jan, produto_mais_ven_fev, 
                                        produto_mais_ven_marco, produto_mais_ven_abril, produto_mais_ven_maio, produto_mais_ven_junho, 
                                        produto_mais_ven_julho, produto_mais_ven_agosto, produto_mais_ven_setembro, 
                                        produto_mais_ven_out, produto_mais_ven_nov, produto_mais_ven_dezem]
        
        print("O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[0][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[0][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[0][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[0][0]["preco_total_de_vendas"], "R$", 
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[1][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[1][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[1][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[1][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[2][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[2][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[2][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[2][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[3][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[3][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[3][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[3][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[4][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[4][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[4][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[4][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[5][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[5][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[5][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[5][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[6][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[6][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[6][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[6][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[7][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[7][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[7][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[7][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[8][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[8][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[8][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[8][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[8][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[8][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[8][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[8][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[9][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[9][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[9][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[9][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[10][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[10][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[10][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[10][0]["preco_total_de_vendas"], "R$",
              "--------------------------------------------------------------",
              "O produto mais vendido no m√™s de",lista_de_todos_mais_vendidos[11][0]["mes"], "foi o -> ", lista_de_todos_mais_vendidos[11][0]["nome_produto"],
              "com ", lista_de_todos_mais_vendidos[11][0]["qtd"], "vendidos, gerando um total de ", lista_de_todos_mais_vendidos[11][0]["preco_total_de_vendas"], "R$")
        return lista_de_todos_mais_vendidos
        

```

### üóÑÔ∏è C√©lula de Banco de Dados

Essa c√©lula cria um banco SQLite3, abre a conex√£o, define a tabela, carrega os dados de um arquivo CSV, aplica as altera√ß√µes e no final fecha a conex√£o.

```python
import sqlite3
import csv

caminho_csv = "data_clean.csv"  # substitua pelo caminho real


conn = sqlite3.connect("dbvendas")
curs = conn.cursor()

curs.execute("""
    CREATE TABLE produtos (
        id INTEGER PRIMARY KEY,
        data TEXT,
        produto_nome TEXT,
        produto_categoria TEXT,
        produto_preco REAL,
        quantidade INTEGER,
        mes TEXT
    )
""")

with open(caminho_csv, newline='', encoding='utf-8') as csvfile:
    leitor = csv.DictReader(csvfile)
    for linha in leitor:
        curs.execute("""
            INSERT INTO produtos (id, data, produto_nome, produto_categoria, produto_preco, quantidade, mes)
            VALUES (?, ?, ?, ?, ?, ?, ?)
        """, (
            int(linha['id']),
            linha['data'],
            linha['produto_nome'],
            linha['produto_categoria'],
            float(linha['produto_preco']),
            int(linha['quantidade']),
            linha['mes']
        ))

conn.commit()


curs.execute("SELECT * FROM produtos")
rows = curs.fetchall()
for row in rows:
    print(row)


conn.close()
```
